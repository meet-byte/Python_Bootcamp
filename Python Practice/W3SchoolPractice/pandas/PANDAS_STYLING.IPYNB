15. Create a dataframe of ten rows, four columns with random values. Write a Pandas program to display bar charts in dataframe on specified columns.



```python
import pandas as pd
import numpy as np
np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
df.iloc[0, 2] = np.nan
df.iloc[3, 3] = np.nan
df.iloc[4, 1] = np.nan
df.iloc[9, 4] = np.nan
print("Original array:")
print(df)
print("\nBar charts in dataframe:")
df.style.bar(subset=['B', 'C'], color='#d65f5f')
```

```python
'''
np.random.seed(24): Sets a seed for reproducibility.
np.linspace(1, 10, 10): Creates an array of 10 evenly spaced values between 1 and 10 for column 'A'.
np.random.randn(10, 4): Generates a 10x4 array of random values from a standard normal distribution (mean=0, std=1).
The resulting DataFrame has 5 columns (A, B, C, D, E), with column 'A' containing 
linearly spaced values and the others containing random values.
'''
import pandas as pd
import numpy as np
import seaborn as sns

np.random.seed(24)
df = pd.DataFrame({'A': np.linspace(1, 10, 10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10, 4), columns=list('BCDE'))],
               axis=1)
print("Original array:")
print(df)
print("\nDataframe - Heatmap style:")

cm = sns.light_palette("red", as_cmap=True)
 
df.style.background_gradient(cmap='viridis') 
```

```python
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4.12345, 5.67891, 6.54321],
    'C': [7, 8, 9]
})

# Apply a style
df.style
```

```python
#  : Introduces the format specification inside the {} placeholder. Everything after the : describes 
#  how the value should be formatted.
df.style.format({'B':"{:.3f}"})
```

```python
print("{:,.2f}".format(1234567.891))
```

```python
df.style.highlight_max()
```

```python
df.style.highlight_min()
```

```python
df.style.highlight_null()
```

```python
df.style.highlight_between(left=0,right=8)
```

```python
def highlight_greater_than_5(value):
    color="red" if value>5 else ""
    return f"background-color:color"

df.style.applymap(highlight_greater_than_5)
```

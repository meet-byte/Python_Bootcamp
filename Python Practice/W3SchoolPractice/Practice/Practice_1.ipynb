```python
from urllib3 import PoolManager,Timeout
http=PoolManager()
timeout=Timeout(connect=2.0,read=4.0)
response=http.request("GET","https://docs.chaicode.com/",timeout=timeout)
print(response.data.decode("utf-8"))
```

# Retry

urllib3 supports automatic retries for failed requests.



```python
from urllib3 import PoolManager,Retry


retry=PoolManager(retries=Retry(total=3,backoff_factor=0.5))
response=retry.request("GET","https://docs.chaicode.com/")

print(response.data.decode("UTF-8"))
```

# 5. Secure Connections (SSL/TLS)

### urllib3 supports HTTPS and can handle secure connections.



```python
pip install certifi
```

```python
import certifi
from urllib3 import PoolManager

http = PoolManager(cert_reqs='CERT_REQUIRED', ca_certs=certifi.where())
response = http.request('GET', 'https://docs.chaicode.com/', headers={'Accept-Encoding': 'gzip'})
print(response.data.decode('utf-8'))
```

```python
import urllib3
import gzip
from io import BytesIO

# Create a PoolManager
http = urllib3.PoolManager()

# Define the URL of the server that sends a compressed response
url = 'https://docs.chaicode.com/'

try:
    # Make a GET request
    response = http.request('GET', url)

    # Check if the response is compressed (content-encoding header)
    if 'gzip' in response.headers.get('content-encoding', ''):
        # Decompress the content manually
        compressed_data = BytesIO(response.data)
        decompressed_data = gzip.GzipFile(fileobj=compressed_data).read()
        print("Decompressed Content:")
        print(decompressed_data.decode('utf-8'))
    else:
        print("Response is not compressed.")

except Exception as e:
    print(f"Error: {e}")
```
